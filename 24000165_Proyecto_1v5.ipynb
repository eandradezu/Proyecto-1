```python
import numpy as np

def regresion_lineal (x,y,epochs, imprimir_error_cada, learning_rate):
    x1=np.array(x)
    x1_1=np.full(len(x1),1)
    x1_array=np.column_stack((x1_1,x1))

    b0=np.full(len(x1),float(1))
    b1=np.full(len(x1),float(1))
    b_array=np.column_stack((b0,b1))

    itx1={}

    for i in range(1,epochs):
        est_y1=np.sum(x1_array* b_array,axis=1)
        dif1=(np.array(y)-est_y1)**2
        err1=np.mean(np.sqrt(dif1))
        if i % imprimir_error_cada == 0:
            print('Iteración ',i, ', Error promedio: ',err1)
        if i % 100 == 0:
            plt.plot(x1, est_y1)
            plt.show()
        itx1[i]=(err1,b_array[0])
        
        errb1=-2*(np.mean(np.array(y)-est_y1))*x1_array
        
        b_array=b_array-(learning_rate*errb1)

    iteracion=list(itx1.keys())
    error=[valor[0] for valor in itx1.values()]
    plt.scatter(iteracion, error)
    print(itx1[max(iteracion)])
    
```
